```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch

model_path = "/content/models/TinyLlama-1.1B-Chat-v1.0"

tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    torch_dtype=torch.bfloat16,
    device_map="auto"   # automatically moves to GPU
)

pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)
print("Loaded TinyLlama (HF format) successfully!")
```

```python
import time, torch, pynvml, pandas as pd, numpy as np, matplotlib.pyplot as plt
from tqdm import tqdm

# initialize NVML (for GPU power measurement)
pynvml.nvmlInit()
handle = pynvml.nvmlDeviceGetHandleByIndex(0)

# define prompts of different lengths
prompt_64   = "What is artificial intelligence?"
prompt_256  = "Explain the history, key methods, and modern advances in artificial intelligence. " * 5
prompt_512  = "Explain the history, key methods, and modern advances in artificial intelligence. " * 10
prompt_1024 = "Explain the history, key methods, and modern advances in artificial intelligence. " * 20

prompts = {64: prompt_64, 256: prompt_256, 512: prompt_512, 1024: prompt_1024}
GT_values = [16, 64, 128]
```

```python
def measure_inference_energy(prompt, max_new_tokens=64):
    torch.cuda.synchronize()
    start_time = time.time()
    start_energy = pynvml.nvmlDeviceGetPowerUsage(handle)  # mW
    energy_accum = 0

    # measure energy sampling loop
    interval = 0.05  # 50 ms sampling
    start = time.time()
    outputs = ""
    while True:
        power = pynvml.nvmlDeviceGetPowerUsage(handle)  # mW
        energy_accum += power * interval  # mW * s = mJ
        time.sleep(interval)
        if time.time() - start > 0.2:  # small warm-up
            break
    # actual generation
    t0 = time.time()
    outputs = pipe(prompt, max_new_tokens=max_new_tokens, do_sample=False)
    torch.cuda.synchronize()
    latency = time.time() - t0
    end_power = pynvml.nvmlDeviceGetPowerUsage(handle)

    # one final read for energy after completion
    energy_accum += end_power * latency
    energy_mJ = energy_accum
    return energy_mJ, latency
```

```python
records = []

for PT, prompt in prompts.items():
    for GT in GT_values:
        for run in range(3):
            print(f"Running PT={PT}, GT={GT}, trial={run+1}")
            E, L = measure_inference_energy(prompt, GT)
            records.append({"PT": PT, "GT": GT, "Energy_mJ": E, "Latency_s": L})

df = pd.DataFrame(records)
df["PT2"] = df["PT"] ** 2
df.to_csv("energy_profile.csv", index=False)
print(df.head())
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

m = LinearRegression().fit(df[["PT2", "GT"]], df["Energy_mJ"])
pred = m.predict(df[["PT2", "GT"]])

mae  = mean_absolute_error(df["Energy_mJ"], pred)
mape = mean_absolute_percentage_error(df["Energy_mJ"], pred)

print(f"Coefficients: a={m.coef_[0]:.4f}, b={m.coef_[1]:.4f}, c={m.intercept_:.2f}")
print(f"MAE={mae:.2f} mJ,  MAPE={mape*100:.2f}%")
```

```python
plt.figure(figsize=(6,4))
for GT in GT_values:
    sub = df[df["GT"]==GT]
    plt.plot(sub["PT"], sub["Energy_mJ"], "o-", label=f"GT={GT}")
plt.xlabel("Prompt Tokens (PT)")
plt.ylabel("Energy (mJ)")
plt.legend()
plt.title("Energy vs PT (fixed GT)")
plt.show()

plt.figure(figsize=(6,4))
for PT in prompts.keys():
    sub = df[df["PT"]==PT]
    plt.plot(sub["GT"], sub["Energy_mJ"], "o-", label=f"PT={PT}")
plt.xlabel("Generated Tokens (GT)")
plt.ylabel("Energy (mJ)")
plt.legend()
plt.title("Energy vs GT (fixed PT)")
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(df["Energy_mJ"], pred, c='r')
plt.xlabel("Measured Energy (mJ)")
plt.ylabel("Predicted Energy (mJ)")
plt.title("Predicted vs Measured Energy")
plt.plot([df["Energy_mJ"].min(), df["Energy_mJ"].max()],
         [df["Energy_mJ"].min(), df["Energy_mJ"].max()], 'k--')
plt.show()
```
