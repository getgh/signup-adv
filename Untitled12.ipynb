```python
from huggingface_hub import snapshot_download
model_path = snapshot_download(
    repo_id="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    local_dir="/content/models/TinyLlama-1.1B-Chat-v1.0"
)
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
model_path = "/content/models/TinyLlama-1.1B-Chat-v1.0"

tokenizer = AutoTokenizer.from_pretrained(model_path)

model = AutoModelForCausalLM.from_pretrained(model_path, device_map="auto")
model.to(device)
```

```python
import torch
import time
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error

torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print("Device:", device)
```

```python
prompts = {
    64: "What is AI?",
    256: "Explain the history, key methods, and modern advances in AI. " * 5,
    512: "Explain the history, key methods, and modern advances in AI. " * 10,
    1024: "Explain the history, key methods, and modern advances in AI. " * 20
}

GT_values = [16, 64, 128]
```

```python
def profile_inference(prompt, max_new_tokens, repeats=3, power_mw=75):
    """
    Approximate energy by latency * power_mw.
    Uses Hugging Face transformers model.
    """
    latencies = []
    energies = []

    for _ in range(repeats):
        inputs = tokenizer(prompt, return_tensors="pt").to(device)
        start = torch.cuda.Event(enable_timing=True)
        end = torch.cuda.Event(enable_timing=True)
        start.record()
        _ = model.generate(**inputs, max_new_tokens=max_new_tokens)
        end.record()
        torch.cuda.synchronize()

        latency = start.elapsed_time(end) / 1000
        energy = latency * power_mw
        latencies.append(latency)
        energies.append(energy)

    return sum(energies)/len(energies), sum(latencies)/len(latencies)
```

```python
results = []

for PT, prompt in prompts.items():
    for GT in GT_values:
        energy, latency = profile_inference(prompt, max_new_tokens=GT, repeats=3)
        results.append({
            "PT": PT,
            "GT": GT,
            "Energy_mJ": energy,
            "Latency_s": latency
        })

df = pd.DataFrame(results)
df['PT2'] = df['PT']**2
df.to_csv("energy_profile.csv", index=False)
df
```

```python
X = df[['PT2', 'GT']]
y = df['Energy_mJ']

model_reg = LinearRegression()
model_reg.fit(X, y)

print("Coefficients:", model_reg.coef_)
print("Intercept:", model_reg.intercept_)

y_pred = model_reg.predict(X)

mae = mean_absolute_error(y, y_pred)
mape = mean_absolute_percentage_error(y, y_pred)
print(f"MAE: {mae:.2f} mJ, MAPE: {mape*100:.2f}%")
```

```python
for GT in GT_values:
    subset = df[df['GT']==GT]
    plt.plot(subset['PT'], subset['Energy_mJ'], marker='o', label=f"GT={GT}")
plt.xlabel("Prompt Tokens (PT)")
plt.ylabel("Energy (mJ)")
plt.title("Energy vs PT for fixed GT")
plt.legend()
plt.show()

for PT in prompts.keys():
    subset = df[df['PT']==PT]
    plt.plot(subset['GT'], subset['Energy_mJ'], marker='o', label=f"PT={PT}")
plt.xlabel("Generated Tokens (GT)")
plt.ylabel("Energy (mJ)")
plt.title("Energy vs GT for fixed PT")
plt.legend()
plt.show()

plt.scatter(y, y_pred)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')
plt.xlabel("Measured Energy (mJ)")
plt.ylabel("Predicted Energy (mJ)")
plt.title("Predicted vs Measured Energy")
plt.show()
```
